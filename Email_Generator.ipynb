```python
import uuid
import chromadb

client = chromadb.PersistentClient('vectorstore')
collection = client.get_or_create_collection(name="portfolio")
```

```python

from sentence_transformers import SentenceTransformer

# Initialize
model = SentenceTransformer('all-MiniLM-L6-v2')

# Step 1: Load the raw CV text
text = document[0].text.strip()

```

```python
print(text)
```

```python
# Step 2: Chunk automatically — split by double newlines
chunks = [chunk.strip() for chunk in text.split('\n\n') if len(chunk.strip()) > 40]
```

```python
chunks
```

```python
# Step 3: Embed & store
for chunk in chunks:
    embedding = model.encode(chunk).tolist()
    collection.add(
        ids=[str(uuid.uuid4())],
        documents=[chunk],
        embeddings=[embedding],
        metadatas=[{"source": "cv", "length": len(chunk)}]
    )
```

```python
# Slightly more streamlined version
relevant_document_parts = []
unique_retrieved_texts = set()

for value in job.values(): # Iterate through values of the job dictionary
    current_queries = []
    if isinstance(value, str):
        if value.strip(): current_queries.append(value.strip())
    elif isinstance(value, list):
        for item in value:
            if isinstance(item, str) and item.strip():
                current_queries.append(item.strip())

    for query_text in current_queries:
        query_results = collection.query(
            query_texts=[query_text],
            n_results=2
        )

        # This check is still highly recommended for robustness:
        retrieved_docs_list = query_results.get('documents')
        if retrieved_docs_list and retrieved_docs_list[0]:
            for doc_text in retrieved_docs_list[0]:
                if doc_text not in unique_retrieved_texts:
                    relevant_document_parts.append(doc_text)
                    unique_retrieved_texts.add(doc_text)

# Printing part remains the same
print("Relevant CV Parts Retrieved Based on Job Elements:")
print("================================================")
if relevant_document_parts:
    for i, part in enumerate(relevant_document_parts):
        print(f"--- Part {i+1} ---\n{part}\n")
else:
    print("No relevant parts found in the CV for the given job elements.")
```

```python
prompt_email = PromptTemplate.from_template(
    """
    ### JOB DESCRIPTION:
    {job_description}

    ### RELEVANT CONTEXT (Illustrative of profiles we can source):
    {cv_extracts}

    ### INSTRUCTION:

    Your task is to write a concise and compelling email to the HR (who posted the job description above)

    Use the "RELEVANT CONTEXT" (which contains the relevent part of my CV) to subtly demonstrate my
    deep understanding of the technical and experiential requirements for this role.
    Do NOT say "we have this candidate" or directly quote large parts of the context.
    Instead, allude to the type of skills, experience, or project involvements (as hinted in the context)
    to show that I am the best candidate.
    The primary goal is to initiate a conversation about how I can be a top-tier candidate for the company.

    Maintain a professional, confident, and helpful tone.
    Remember you are Shihab Hassan Naim.
    The output should be only the email body, with no extra titles or preambles.

    ### EMAIL:
    """
)


chain_email = prompt_email | llm


res = chain_email.invoke({
    "job_description": str(job),  # 'job' is the dictionary from json_res
    "cv_extracts": relevant_document_parts
})

# Print the generated email content
print("Generated Email:")
print("----------------------------------------------------")
print(res.content)
```
